```python
# Connect to the database
conn = sqlite3.connect('transactions.db')

# Query the database
cur = conn.cursor()

sql = "select day, day_total from (select date(transaction_time) as day, sum(transaction_amount) as day_total from transactions group by day order by day)"

cur.execute(sql)

# Fetch the results
rows = cur.fetchall()

# Print the results
for row in rows:
  print(row)

# Close the connection
conn.close()
```

```python
# Connect to the database
conn = sqlite3.connect('transactions.db')

# Query the database
cur = conn.cursor()

sql = "select * from (select day, avg(day_total) OVER (Order BY day ROWS BETWEEN 2 PRECEDING AND CURRENT ROW) as moving_avg \
from (select date(transaction_time) as day, sum(transaction_amount) as day_total from transactions group by day order by day)) where day = '2021-01-31'"

cur.execute(sql)

# Fetch the results
rows = cur.fetchall()

# Print the results
for row in rows:
  print(row)

# Close the connection
conn.close()
```
